# MS Word Print-outs

```python
import pandas as pd
from docx import Document
from docx.shared import Pt, RGBColor, Inches
from docx.enum.text import WD_ALIGN_PARAGRAPH


def create_printable_cheat_sheet(df, output_path="./printouts/Master_Scripts_Cheat_Sheet.docx"):
    # 1. Initialize Document
    doc = Document()

    # Optional: Adjust margins to fit more on a printed page
    sections = doc.sections
    for section in sections:
        section.top_margin = Inches(0.75)
        section.bottom_margin = Inches(0.75)
        section.left_margin = Inches(0.75)
        section.right_margin = Inches(0.75)

    # 2. Add Main Title
    title = doc.add_heading("Dating Scripts: Qualitative Codebook", 0)
    title.alignment = WD_ALIGN_PARAGRAPH.CENTER

    subtitle = doc.add_paragraph("Master Scripts & Core Logics Reference Sheet")
    subtitle.alignment = WD_ALIGN_PARAGRAPH.CENTER

    doc.add_paragraph("_" * 50)  # Visual separator

    # 3. Iterate through the Master Clusters
    for index, row in df.iterrows():
        # Handle the name
        script_name = str(row.get('master_cluster_name', f"Cluster {row.get('Master_Cluster_ID', index)}"))

        # Add Script Heading (Level 2 looks good for printing)
        h = doc.add_heading(script_name, level=2)
        h.style.font.color.rgb = RGBColor(0x1F, 0x4E, 0x79)  # Dark blue for readability

        # Process the Core Logic
        raw_logic = str(row.get('core_logic', ''))

        # Split by the pipe character you used during aggregation
        logics_list = [logic.strip() for logic in raw_logic.split('|') if logic.strip()]

        # Keep only the top 3 (or fewer if there aren't 3)
        top_logics = logics_list[:3]

        if not top_logics:
            doc.add_paragraph("No core logic recorded.", style='List Bullet')
        else:
            for logic in top_logics:
                # Add as a bullet point
                p = doc.add_paragraph(logic, style='List Bullet')
                # Optional: make the font slightly smaller to save space
                for run in p.runs:
                    run.font.size = Pt(11)

        # Add a tiny bit of space after each script block
        doc.add_paragraph()

    # 4. Save the document
    doc.save(output_path)
    print(f"✅ Printable Cheat Sheet saved to: {output_path}")


# --- EXECUTE ---
# Assuming your dataframe is named 'master_clusters_df' or 'new_master_clusters_df'
create_printable_cheat_sheet(master_clusters_df)
```
